```python
import numpy as np
import matplotlib.pyplot as plt
n = 20
err = []
err_2 = []
h = []
vals = []
for i in range(n):
    h.append(10**(-i))
    err.append(abs(np.cos(1.5) - ((np.sin(1.5 + 10**(-i)) - np.sin(1.5)) / 10 ** (-i))))
    err_2.append(abs(np.cos(1.5) - ((2 * np.cos(0.5 * (3 + 10**(-i))) * np.sin(0.5 *(10**(-i)))) / 10 ** (-i))))
    
plt.xscale("log")
plt.yscale("log")

plt.xlabel("h")
plt.ylabel("absolute error")

plt.plot(h, err, label = 'Given formula')
plt.plot(h, err_2, label = 'Modified')

plt.legend()
```
