```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

x = np.arange(0, 5.5, 0.5)
y = np.array([0.2, 2, 3.97, 4.22, 5.99, 6.23, 5.43, 4.01, 3.73, 4.99, 5.55], dtype = float)

A = np.vstack([np.ones_like(x), x, x ** 2, x ** 3, x ** 4, x ** 5]).T

N = np.dot(np.transpose(A), A)
z = np.dot(np.transpose(A), y)

w = la.solve(N, z)
print(w)

def poly(t, w):
    n = len(w)
    z = 0
    for i in range(n):
        z += w[i] * (t ** i)
    return z

t = np.linspace(0, 5, 200, endpoint = True)
f = poly(t, w)

plt.plot(t, f)
plt.ylim(-1, 7)
plt.scatter(x, y)
```
