```python
import numpy as np

def newton_method(df, d2f, x0, tol, max_iter):
    x = x0
    x_new = x - df(x) / d2f(x)
    count = 0
    while ((count < max_iter) and (abs(x_new - x) >= tol)):
        count += 1
        x = x_new
        x_new = x - df(x) / d2f(x)
        
    return x_new, count
```

```python
def df(x):
    return((2 * x ** 2 - 1) * np.exp(-x ** 2))

def d2f(x):
    return(2 * x * (3 - 2 * x ** 2) * np.exp(-x ** 2))

root1, count1 = newton_method(df, d2f, 1, 1.0e-5, 100)
print("Root : ", root1)
print("Count : ", count1)
print(df(root1))
```

```python
def df(x):
    return(2 * x - 4 * np.sin(x))

def d2f(x):
    return(2 - 4 * np.cos(x))

root2, count2 = newton_method(df, d2f, 1.5, 1.0e-10, 100)
print("Root : ", root2)
print("Count : ", count2)
print(df(root2))
```

```python
def df(x):
    return(4 * x ** 3 - 42 * x ** 2 + 120 * x - 70)

def d2f(x):
    return(12 * x ** 2 - 84 * x + 120)

root3, count3 = newton_method(df, d2f, 1.5, 1.0e-10, 100)
print("Root : ", root3)
print("Count : ", count3)
print(df(root3))
```
